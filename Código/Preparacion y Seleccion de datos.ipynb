```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importamos los datos
datos_estacion_orig = pd.read_excel("../Datos/datos_estacion_temporal.xlsx");
```

```python
#Obtener la forma de los datos (Columnas y filas)
datos_estacion_orig.shape
```

```python
#Mostrar informacion de los datos
datos_estacion_orig.info()
```

```python
#Creación de la columna "Fecha"

datos_estacion_con_fecha = datos_estacion_orig.astype({'gestion': str,'mes': str,'dia':str})

datos_estacion_con_fecha['Fecha'] = datos_estacion_con_fecha['gestion']+'-'+datos_estacion_con_fecha['mes']+'-'+datos_estacion_con_fecha['dia']
datos_estacion_con_fecha['Fecha'] = pd.to_datetime(datos_estacion_con_fecha['Fecha'])

datos_estacion_con_fecha.set_index('Fecha',inplace=True)

datos_estacion_con_fecha.info()
```

```python
# Contar valores nulos por columna
datos_estacion_con_fecha.isnull().sum()
```

```python
# Eliminar filas con valores nulos a excepción de los de la columna "ETo"
columnas_con_datos_nulos = ['Presion','"Velocidad de Viento Media"','"Velocidad de Viento Máxima"','"Velocidad de Viento Mínima"','"Direccion de Viento Media"','"Direccion de Viento Máxima"','"Direccion de Viento Mínima"','Precipitación']
datos_estacion_noNaN = datos_estacion_con_fecha.dropna(subset=columnas_con_datos_nulos)
datos_estacion_noNaN.isnull().sum()
```

```python
#Mostrar cantidad de filas y columnas tras la eliminación de nulos (Sin contar con ETo)
print("Cantidad de Filas y Columnas: ",datos_estacion_noNaN.shape)
```

```python
# Mostrar la distribución con boxplots de las variables de temperatura
plt.figure(figsize=(7,6))
plt.title("Distribución de los datos relacionados a la Temperatura")
plt.ylabel("Grado Centígrado (C°)")

paleta = ['#ff2d00','#118DFF','#12239e']
sns.set_palette(paleta)
sns.boxplot(data=datos_estacion_noNaN[['"Temperatura Máxima"','"Temperatura Media"','"Temperatura Mínima"']])
```

```python
# Mostrar la distribución con boxplots de las variables de humedad relativa

plt.figure(figsize=(8,6))
plt.title("Distribución de datos relacionados a la Humedad Relativa")
plt.ylabel("Porcentaje (%)")

paleta = ['#ff2d00','#118DFF','#12239e']
sns.set_palette(paleta)
sns.boxplot(data=datos_estacion_noNaN[['"Humedad Relativa Máxima"','"Humedad Relativa Media"','"Humedad Relativa Mínima"']])
```

```python
#Eliminar valores erroneos de las columnas de temperatura minima, media y humedad relativa media (valores minimos que superan al máximo)

condicion_sin_TempMin_sobre_TempMed = datos_estacion_noNaN['"Temperatura Mínima"'] > datos_estacion_noNaN['"Temperatura Media"']
datos_estacion_noNaN_noErr1 = datos_estacion_noNaN[~condicion_sin_TempMin_sobre_TempMed]

condicion_sin_TempMed_sobre_TempMax = datos_estacion_noNaN_noErr1['"Temperatura Media"'] > datos_estacion_noNaN_noErr1['"Temperatura Máxima"']
datos_estacion_noNaN_noErr2 = datos_estacion_noNaN_noErr1[~condicion_sin_TempMed_sobre_TempMax]

condicion_sin_HumMed_sobre_HumMax = datos_estacion_noNaN_noErr2['"Humedad Relativa Media"'] > datos_estacion_noNaN_noErr2['"Humedad Relativa Máxima"']
datos_estacion_noNaN_noErr3 = datos_estacion_noNaN_noErr2[~condicion_sin_HumMed_sobre_HumMax]

datos_estacion_noNaN_noErr3.shape
```

```python
# Seleccionamos las variables cuantitativas para mostrar estadísticos

datos_estacion_varMeteo = datos_estacion_noNaN_noErr3[['"Temperatura Máxima"','"Temperatura Media"','"Temperatura Mínima"','"Humedad Relativa Máxima"',
                                       '"Humedad Relativa Media"','"Humedad Relativa Mínima"','Presion','"Presion Media"','"Velocidad de Viento Media"',
                                       '"Presion Máxima"','"Presion Mínima"','"Velocidad de Viento Máxima"','"Velocidad de Viento Mínima"',
                                       '"Direccion de Viento Media"','"Direccion de Viento Máxima"','"Direccion de Viento Mínima"','ETo','Precipitación']]
datos_estacion_varMeteo.describe().transpose()
```

```python
# Eliminacion de valores atópicos, utilizando rango intercuantilico (IQR)

datos_estacion_sin_atipicos = datos_estacion_varMeteo.copy()
print("Con Valores Atípicos: ",datos_estacion_sin_atipicos.shape)

def eliminar_atipicos(datos,columnas):
    datos_sin_atipicos = datos.copy()
    
    for columna in columnas:
        
        # Eliminamos valores que no esten dentro del 1% y 99% de los datos 
        Q1 = datos[columna].quantile(0.01)
        Q3 = datos[columna].quantile(0.99)
        IQR = Q3 - Q1
        
        limite_inf = Q1 - 1.5 * IQR
        limite_sup = Q3 + 1.5 * IQR
        
        datos_sin_atipicos = datos_sin_atipicos[
            (datos_sin_atipicos[columna] >= limite_inf) &
            (datos_sin_atipicos[columna] <= limite_sup)
        ]
        
    return datos_sin_atipicos

columnas_a_limpiar_ati = [
    'Presion','"Presion Media"','"Velocidad de Viento Media"',
    '"Presion Máxima"','"Presion Mínima"','"Velocidad de Viento Máxima"','"Velocidad de Viento Mínima"',
    '"Direccion de Viento Media"'
]
datos_estacion_sin_atipicos = eliminar_atipicos(datos_estacion_sin_atipicos, columnas_a_limpiar_ati)
print("Sin Valores Atípicos: ",datos_estacion_sin_atipicos.shape)
```

```python
#Obtenemos estadisticos de los datos sin valores atípicos extremos
datos_estacion_sin_atipicos.describe().transpose()
```

```python
print("Sin Valores Atípicos: ",datos_estacion_sin_atipicos.shape)
```

```python
# Creamos un mapa de calor para ver las correlaciones entre las variables meteorológicas

datos_estacion_sin_atipicos_corr = datos_estacion_sin_atipicos.corr(method='pearson')

plt.figure(figsize=(12, 8))
plt.title("Correlación entre Variables Meteorológicas", fontsize=16)
sns.heatmap(datos_estacion_sin_atipicos_corr, annot=True, fmt=".2f")
plt.show()
```
